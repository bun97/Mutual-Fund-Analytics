```python
import pandas as pd

fund_master = pd.read_csv(r"E:\Mutual Fund Analytics\data\raw\01_fund_master.csv")

# See unique fund house
print(fund_master['fund_house'].unique())

# See unique category
print(fund_master['category'].unique())

# See unique sub_category
print(fund_master['sub_category'].unique())

# See unique risk category
print(fund_master['risk_category'].unique())

# Load both files

fund_master = pd.read_csv(r"E:\Mutual Fund Analytics\data\raw\01_fund_master.csv")
nav_history = pd.read_csv(r"E:\Mutual Fund Analytics\data\raw\02_nav_history.csv")

# Get unique codes from each files
codes_in_master = set(fund_master['amfi_code'])
codes_in_nav = set(nav_history['amfi_code'])

#Find codes that are in master but NOT in nav history
missing = codes_in_master - codes_in_nav

# Print the data quality report
print(f"Total funds in master: {len(codes_in_master)}")
print(f"Total funds in NAV history: {len(codes_in_nav)}")
print(f"Missing codes: {len(missing)}")
print(missing)
```
